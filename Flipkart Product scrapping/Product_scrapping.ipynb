```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
name_list = []
price_list = []
ratings = []
reviews_list = []

def name(soup, name_list):
    for item in soup.find_all("div", {"class": "KzDlHZ"}):
        name_list.append(item.text.strip())

def price(soup, price_list):
    for item in soup.find_all("div", {"class": "Nx9bqj _4b5DiR"}):
        price_list.append(item.text.strip())

def rating(soup, ratings):
    for item in soup.find_all("span", {"class": "Y1HWO0"}):
        ratings.append(item.text.strip())

def reviews(soup, reviews_list):
    for item in soup.find_all("span",{"class":"Wphh3N"}):
        reviews_list.append(item.text.strip())

url = "https://www.flipkart.com/search?q=mobiles"
headers = {"User-Agent": "Mozilla/5.0"}

try:
    while True:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, "lxml")

        name(soup, name_list)
        price(soup, price_list)
        rating(soup, ratings)
        reviews(soup, reviews_list)

        # Find all pagination links
        next_button = None
        for link in soup.find_all("a", {"class": "_9QVEpD"}):
            span = link.find("span")
            if span and span.text.strip().lower() == "next":
                next_button = link
                break

        if not next_button or not next_button.get("href"):
            print("No more pages found.")
            break
            
        print("visiting:",url)

        next_link = "https://www.flipkart.com" + next_button.get("href")
        url = next_link # Politeness delay
        # time.sleep(0.5)
except Exception as e:
    print("Error occurred:", e)

print("Names:", len(name_list))
print("Prices:", len(price_list))
print("Ratings:", len(ratings))
print("Reviews:", len(reviews_list))
```

```python
df = pd.DataFrame({"Product Name":pd.Series(name_list), "Price":pd.Series(price_list), "Ratings":pd.Series(ratings), "Reviews":pd.Series(reviews_list)})
```
